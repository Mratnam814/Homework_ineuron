9 . Average weekly response time for each agent

```python
df
```

```python
def total_seconds(seconds):
    hh,mm,ss = seconds.split(':')
    return int(hh) * 3600 + int(mm) * 60 + int(ss)
   
```

```python
response_time=[]
for i in df['Average Resolution Time']:
    response_time.append(total_seconds(i))
response_time
df['response-time']=response_time
    
```

```python
df
```

```python
df.groupby(['Agent Name','week'])['response-time'].mean()
```

10 . Average weekely resolution time for each agents

```python
df
```

```python
def resolution(res):
    hh,mm,ss = res.split(':')
    return int(hh)*3600+int(mm)*60+int(ss)
```

```python
resolution_time=[]
for i in df['Average Resolution Time']:
    resolution_time.append(resolution(i))
resolution_time
df['resolution-time']=resolution_time
```

```python
df
```

```python
df.groupby(['Agent Name','week'])['resolution-time'].mean()
```
